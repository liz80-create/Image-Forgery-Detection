```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

    def change_dimensions(self, new_length, new_width):
        self.length = new_length
        self.width = new_width

    def report_dimensions(self):
        return f"Length: {self.length}, Width: {self.width}"


def rectangle_menu(rectangle):
    while True:
        print("\nRectangle Menu:")
        print("1. Display Dimensions")
        print("2. Calculate Area")
        print("3. Calculate Perimeter")
        print("4. Change Dimensions")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            print("Rectangle Dimensions:", rectangle.report_dimensions())
        elif choice == '2':
            print("Area:", rectangle.area())
        elif choice == '3':
            print("Perimeter:", rectangle.perimeter())
        elif choice == '4':
            new_length = float(input("Enter new length: "))
            new_width = float(input("Enter new width: "))
            rectangle.change_dimensions(new_length, new_width)
            print("Dimensions changed.")
        elif choice == '5':
            break
        else:
            print("Invalid choice. Please try again.")


length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

rectangle = Rectangle(length, width)

rectangle_menu(rectangle)
```

```python
class Polygon:
    def __init__(self, sides):
        self.sides = sides

    def area(self):
        pass

    def perimeter(self):
        pass

    def change_dimensions(self, new_dimensions):
        pass

    def report_dimensions(self):
        pass


class Rectangle(Polygon):
    def __init__(self, length, width):
        super().__init__(sides=4)
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

    def change_dimensions(self, new_length, new_width):
        self.length = new_length
        self.width = new_width

    def report_dimensions(self):
        return f"Length: {self.length}, Width: {self.width}"


class Square(Rectangle):
    def __init__(self, side):
        super().__init__(length=side, width=side)

    def change_dimensions(self, new_side):
        super().change_dimensions(new_side, new_side)


def main():
    while True:
        print("\nMenu:")
        print("1. Create Rectangle")
        print("2. Create Square")
        print("3. Display Dimensions")
        print("4. Calculate Area")
        print("5. Calculate Perimeter")
        print("6. Change Dimensions")
        print("0. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            length = float(input("Enter length of rectangle: "))
            width = float(input("Enter width of rectangle: "))
            rectangle = Rectangle(length, width)
            print("Rectangle created.")

        elif choice == "2":
            side = float(input("Enter side length of square: "))
            square = Square(side)
            print("Square created.")

        elif choice == "3":
            try:
                print("Rectangle Dimensions:", rectangle.report_dimensions())
            except NameError:
                print("Rectangle not created yet.")

            try:
                print("Square Dimensions:", square.report_dimensions())
            except NameError:
                print("Square not created yet.")

        elif choice == "4":
            try:
                print("Rectangle Area:", rectangle.area())
            except NameError:
                print("Rectangle not created yet.")

            try:
                print("Square Area:", square.area())
            except NameError:
                print("Square not created yet.")

        elif choice == "5":
            try:
                print("Rectangle Perimeter:", rectangle.perimeter())
            except NameError:
                print("Rectangle not created yet.")

            try:
                print("Square Perimeter:", square.perimeter())
            except NameError:
                print("Square not created yet.")

        elif choice == "6":
            try:
                new_length = float(input("Enter new length for rectangle: "))
                new_width = float(input("Enter new width for rectangle: "))
                rectangle.change_dimensions(new_length, new_width)
                print("Rectangle dimensions updated.")
            except NameError:
                print("Rectangle not created yet.")

            try:
                new_side = float(input("Enter new side length for square: "))
                square.change_dimensions(new_side)
                print("Square dimensions updated.")
            except NameError:
                print("Square not created yet.")

        elif choice == "0":
            print("Exiting the program.")
            break

        else:
            print("Invalid choice. Please enter a valid option.")


main()
```

```python
import tkinter as tk
import math

class DrawingApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Drawing Application")
        self.canvas = tk.Canvas(self.master, width=400, height=400, bg="pink")
        self.canvas.pack(fill=tk.BOTH, expand=True)
        self.canvas.bind("<Button-1>", self.start_shape)
        self.canvas.bind("<B1-Motion>", self.draw_shape)
        self.start_x = None
        self.start_y = None
        self.shapes = []

        rectangle_button = tk.Button(self.master, text="Rectangle", command=self.set_rectangle)
        rectangle_button.pack(side=tk.LEFT)

        triangle_button = tk.Button(self.master, text="Triangle", command=self.set_triangle)
        triangle_button.pack(side=tk.LEFT)

        circle_button = tk.Button(self.master, text="Circle", command=self.set_circle)
        circle_button.pack(side=tk.LEFT)

        hexagon_button = tk.Button(self.master, text="Hexagon", command=self.set_hexagon)
        hexagon_button.pack(side=tk.LEFT)

        star_button = tk.Button(self.master, text="Star", command=self.set_star)
        star_button.pack(side=tk.LEFT)

        clear_button = tk.Button(self.master, text="Eraser", command=self.clear_canvas)
        clear_button.pack()

    def set_rectangle(self):
        self.current_shape = "rectangle"

    def set_triangle(self):
        self.current_shape = "triangle"

    def set_circle(self):
        self.current_shape = "circle"

    def set_hexagon(self):
        self.current_shape = "hexagon"

    def set_star(self):
        self.current_shape = "star"

    def start_shape(self, event):
        self.start_x = event.x
        self.start_y = event.y

    def draw_shape(self, event):
        if self.start_x is not None and self.start_y is not None:
            x, y = event.x, event.y
            if self.current_shape == "rectangle":
                self.draw_rectangle(x, y)
            elif self.current_shape == "triangle":
                self.draw_triangle(x, y)
            elif self.current_shape == "circle":
                self.draw_circle(x, y)
            elif self.current_shape == "hexagon":
                self.draw_hexagon(x, y)
            elif self.current_shape == "star":
                self.draw_star(x, y)

    def draw_rectangle(self, x, y):
        self.shapes.append(('rectangle', (self.start_x, self.start_y, x, y)))
        self.canvas.create_rectangle(self.start_x, self.start_y, x, y, outline="black", fill="red")

    def draw_triangle(self, x, y):
        x3 = self.start_x + (x - self.start_x) / 2
        y3 = self.start_y
        self.shapes.append(('triangle', (self.start_x, self.start_y, x, y, x3, y3)))
        self.canvas.create_polygon(self.start_x, self.start_y, x, y, x3, y3, outline="black", fill="blue")

    def draw_circle(self, x, y):
        radius = ((x - self.start_x) ** 2 + (y - self.start_y) ** 2) ** 0.5
        self.shapes.append(('circle', (self.start_x, self.start_y, radius)))
        self.canvas.create_oval(self.start_x - radius, self.start_y - radius, self.start_x + radius,
                                 self.start_y + radius, outline="black", fill="green")

    def draw_hexagon(self, x, y):
        radius = ((x - self.start_x) ** 2 + (y - self.start_y) ** 2) ** 0.5
        angles = [angle * 2 * math.pi / 6 for angle in range(6)]
        points = [(self.start_x + radius * math.cos(angle), self.start_y + radius * math.sin(angle)) for angle in angles]
        self.shapes.append(('hexagon', points))
        self.canvas.create_polygon(points, outline="black", fill="purple")

    def draw_star(self, x, y):
        radius = ((x - self.start_x) ** 2 + (y - self.start_y) ** 2) ** 0.5
        angles = [angle * 2 * math.pi / 10 for angle in range(10)]
        points = [(self.start_x + radius * math.cos(angle), self.start_y + radius * math.sin(angle)) for angle in angles[::2]]
        points += [(self.start_x + radius * 0.5 * math.cos(angle), self.start_y + radius * 0.5 * math.sin(angle)) for angle in angles[1::2]]
        self.shapes.append(('star', points))
        self.canvas.create_polygon(points, outline="black", fill="orange")

    def clear_canvas(self):
        self.canvas.delete("all")
        self.shapes = []

def main():
    root = tk.Tk()
    app = DrawingApp(root)
    root.geometry("+800+400")
    root.mainloop()

if __name__ == "__main__":
    main()
```
